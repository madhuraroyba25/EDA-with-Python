```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the data into a Pandas dataframe
df = pd.read_csv("data/kc_house_data_original.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking if anything is missing and needs cleaning
df.isnull().sum()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
#Scatterplot
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x="sqft_living",
    y="price",
    hue="bedrooms"
)

plt.title("Price vs Living Area (Colored by Bedrooms)")
plt.xlabel("Living Area (sq. ft.)")
plt.ylabel("Price")

plt.show()
```

The plot shows a positive relationship between living area and price. Most observations are clustered at smaller sizes, with a few larger homes having higher prices.

```python
#Faceted Histogram
sns.displot(
    data=df,
    x="price",
    col="waterfront",
    bins=20,
    log_scale=True
)

plt.show()
```

The price distribution differs between waterfront and non-waterfront homes. Waterfront properties are fewer and tend to have higher prices. A log scale is used to visualize the wide range of prices in a better way.

```python
df["sale_month"] = df["date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("sale_month").size().reset_index(name="num_sales")
```

```python
monthly_sales["sale_month"] = monthly_sales["sale_month"].dt.to_timestamp()
```

```python
df[["date", "sale_month"]].head()
```

```python
#Time Series Plot
sns.lineplot(
    data=monthly_sales,
    x="sale_month",
    y="num_sales"
)

plt.title("Number of Homes Sold per Month")
plt.xlabel("Month")
plt.ylabel("Number of Sales")

plt.show()
```

The plot shows the number of homes sold per month over time. Sales increase during mid-2014, decline towards the end of the year, and then rise again in early 2015.
